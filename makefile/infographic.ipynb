```python
import matplotlib.pyplot as plt 
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.patches import Rectangle
```

```python
# creating title 
font_path = '../data/econosans/ecosans-bol.woff2.ttf'  # Update this path
custom_font = FontProperties(fname=font_path, size=35)
font_path2 = '../data/econosans/EconoSansReduced-35Thin.ttf' 
custom_font2 = FontProperties(fname=font_path2, size=100)

fig, ax = plt.subplots(figsize=(24, 1))
fig.text(
    0.38, 0.5, "Migration Flows in Italy ",
    fontsize=40, fontweight="bold", fontproperties=custom_font
)
fig.text(
    0.18, 0.1, "Data on Migrants'Arrivals in Italy independently collected by the project LiberiamoliTutti that releases a dataset every month of public utility",
    fontsize=25, fontproperties=custom_font2, fontweight= "bold"
)
ax.set_axis_off()
plt.savefig('../outputs/title', bbox_inches='tight', dpi=300, facecolor="white")
plt.show()
```

```python


fig = plt.figure(figsize=(10, 4))  # Set the figure size as requested
ax = fig.add_subplot(111)

# General introduction text and keywords with descriptions
intro_text = "Different centers present:"
keywords_descriptions = [
    ("HOTSPOTS:", "crisis centers located at landing areas providing initial medical aid", '#006BA2'),
    ("ACCOGLIENZA CENTERS:", "focused on asylum seekers offering healthcare and social services", '#3EBCD2'),
    ("SAI (Reception and Integration System)", "serves those with protection status focusing on employment and integration", '#379A8B'),
    ("SIPROIMI now replaced by SAI:", "for protected individuals or special categories excluding asylum seekers", '#EBB434')
]

# Initial Y position and line height for manual adjustments
y_pos = 0.85
line_height = 0.15  # Adjust for the new figure dimensions

# Plot each piece of text
fig.text(0.05, y_pos, intro_text, fontsize=20, fontproperties=custom_font)
y_pos -= line_height  # Adjust position after the intro text
x = [0.013, 0.013, 0.01, 0.012]
i = 0
for keyword, description, color in keywords_descriptions:
    fig.text(0.05, y_pos, keyword, fontsize=15, fontproperties=custom_font, color=color)
    keyword_width_approx = len(keyword) * x[i] # Adjust this as necessary
    fig.text(0.05 + keyword_width_approx, y_pos, description, fontsize=14, fontproperties=custom_font2)
    y_pos -= line_height
    i +=1

# Draw a rectangle around all the text
# Adjust these values as needed to fit all your text
rect_x = 0.03  # X position of the rectangle
rect_y = y_pos + line_height/2  # Y position of the rectangle, adjusted to the bottom text
rect_width = 0.94  # Width of the rectangle
rect_height = 0.85 - rect_y + 0.02  # Height of the rectangle

# Create a rectangle patch
rect = Rectangle((rect_x, rect_y), rect_width, rect_height, linewidth=1, edgecolor='red', facecolor='none')
ax.add_patch(rect)
ax.text(x=0.12, y=0.1, s="""Source: Cruscotto statistico giornaliero via www.libertaciviliimmigrazione.dlci.interno.gov.it/it \n [LiberiamoliTutti!](/https://github.com/ondata/liberiamoli-tutti/tree/main/sbarchi-migranti)""", transform=fig.transFigure, ha='left', fontsize=15, alpha=.7, fontproperties=custom_font2)
plt.axis('off')  # Hide axes
plt.tight_layout(pad=1)  # Adjust the padding
plt.savefig('../outputs/description', bbox_inches='tight', dpi=300, facecolor="white")
plt.show()

```

```python
from PIL import Image

# Define the DPI for the conversion (common print quality is 300 DPI)
dpi = 300

# Function to convert inches to pixels
def inches_to_pixels(inches):
    return int(inches * dpi)

# Load the images
italian_map = Image.open('../outputs/italian_map.png')

prova_title = Image.open('../outputs/title.png')
world_map = Image.open('../outputs/world_map.png')
pies = Image.open('../outputs/pies.png')
timeserie = Image.open('../outputs/timeserie.png')
description = Image.open('../outputs/description.png')




# Calculate the dimensions based on the DPI
width_A = inches_to_pixels(10)  # Width of Italian map in pixels
height_A = inches_to_pixels(15)  # Height of Italian map in pixels

width_B = inches_to_pixels(24)   # Width of the title in pixels
height_B = inches_to_pixels(1)  # Height of the title in pixels

width_C = inches_to_pixels(14) # Width of World map in pixels
height_C = inches_to_pixels(9.5)  # Height of World map in pixels

width_D = inches_to_pixels(14) # Width of Timeseries in pixels
height_D = inches_to_pixels(5)  # Height of Timeseries in pixels

# Assuming pies.png is the combined pie charts image
width_F = inches_to_pixels(14)  # Width of pies in pixels
height_F = inches_to_pixels(4.5)  # Height of pies in pixels

width_X = inches_to_pixels(10)  # Width of description
height_X = inches_to_pixels(4)  # Height of description



# Create a new image with the intended final size in pixels
final_width = inches_to_pixels(24)
final_height = inches_to_pixels(20)
final_image = Image.new('RGB', (final_width, final_height), 'white')

# Resize images if necessary
italian_map = italian_map.resize((width_A, height_A))
prova_title = prova_title.resize((width_B, height_B))
world_map = world_map.resize((width_C, height_C))
timeserie = timeserie.resize((width_D, height_D))
pies = pies.resize((width_F, height_F))
description = description.resize((width_X, height_X))

# Paste the images into the final image at calculated positions
final_image.paste(prova_title, (0, 0))  # title
final_image.paste(italian_map, (0,height_B))
final_image.paste(world_map, (width_A, height_B))  # C to the right of A
final_image.paste(timeserie, (width_A, height_B + height_C))  # D below C
final_image.paste(pies, (width_A, height_B + height_C + height_D )) 
final_image.paste(description, (0, height_B + height_A )) 

# Save the final image to a file
final_image_path = '../outputs/final_combined_image.png'
final_image.save(final_image_path)

# Print out the path to the saved image
print(f"Image saved to {final_image_path}")
```
